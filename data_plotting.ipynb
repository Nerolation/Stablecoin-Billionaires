```python
idx = pd.date_range('28/11/2017', '30/06/2020')
_all = {}
for i in _all_transfer:
    token = i.split('/')[1]
    df = pd.read_csv(i, index_col='dates').drop('Unnamed: 0', axis=1)
    df.index = pd.DatetimeIndex(df.index)
    _all[token] = df.reindex(idx, fill_value=0).loc['28/11/2017':,:]
_all['tether']['usdc']=_all['usdc']
_all['tether']['paxos']=_all['paxos']
_all['tether']['trueusd']=_all['trueusd']
_all['tether']['dai']=_all['dai']
_all['tether']['binanceusd']=_all['binanceusd']
_all['tether']['husd']=_all['husd']

df = _all['tether']
```

```python
df['sum'] = df['txs']+df['trueusd']+df['usdc']+df['dai']+df['paxos']+df['binanceusd']+df['husd']
df['cum_tether'] = df['txs']/df['sum']
df['cum_trueusd'] = df['trueusd']/df['sum']
df['cum_usdc'] = df['usdc']/df['sum']
df['cum_paxos'] = df['paxos']/df['sum']
df['cum_dai'] = df['dai']/df['sum']
df['cum_binanceusd'] = df['binanceusd']/df['sum']
df['cum_husd'] = df['husd']/df['sum']
df.fillna(0)
df = df.loc['2019-01-01':,]
```

```python
sec_week = 60*60*24*7
tok_gas = {}
for i in _all_tx:
    tk = i.split('/')[1]
    ts = 1577836800 # 01-01-2020
    df = pd.read_csv(i, usecols=['timestamp','gas_price','gas_used'])
    df['costs'] = df['gas_price']*df['gas_used']/10**18
    week = 0
    weeks = {}
    while ts < 1593561600:
        _df = df[(df['timestamp'] >= ts) & (df['timestamp'] < ts+sec_week)]
        weeks[week] = _df['costs'].mean()
        week += 1
        ts += sec_week
    tok_gas[tk] = weeks
```

```python
def getlabel(i):
        global counter
        if i:
            if i != 'others':
                label.append(i + '...')
            else:
                label.append(i)
        else:
            label.append('')
        counter += 1

addr = {}
dfs = {}

for i in _all_positive_cumulated_balances:
    name = i.split('/')[1]
    if name in ['husd', 'binanceusd']:
        slices = 10 if name == 'binanceusd' else 10
        labels = 5 if name == 'binanceusd' else 2
    else:
        slices = 40
        labels = 8
    df = pd.read_csv(i, index_col='Unnamed: 0')
    aa = df.iloc[df.index[-1]-slices:]
    bb = df['balance'].iloc[:df.index[-1]-slices]
    df = aa.append(pd.DataFrame({'address': 'others', 'balance': sum(bb)}, index=[0]))
    label = []
    counter = 0
    
    
    [getlabel(i[:5]) if counter >= len(df)-labels else getlabel('') for i in df['address']  ]
    addr[name] = label
    dfs[name] = df
    label = []
    print()
```
